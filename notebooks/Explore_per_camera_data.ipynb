```python
from IPython.display import Image 

def look_at_random_animal_from_web(specie, random_state, width=500):
    capture_ids = df_full_annotations.loc[df_full_annotations.question__species == specie].capture_id
    if capture_ids.size == 0:
        return 'No specie matching the description'
    filenames = df_full_images.loc[capture_ids].file_name
    filename = filenames.sample(random_state=random_state).values[0]
    print(filename)
    return Image(url= str(filename), width=500, height=500)
```
