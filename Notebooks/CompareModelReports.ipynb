```python
import pandas as pd
import os
```

```python
# Inputs
results_dir = r"C:\Users\mfarj\Documents\projects\aigoeswild\results"
rusmodel_report = r"res18_adam_epoch1_rus_phase1_15_test_CR.csv"
rosrusmodel_report = r"res18_adam_epoch1_ros_rus_phase1_14_test_CR.csv"

fn_combined = r"rusmodel_report_vs_rosrusmodel_report.csv"
```

```python
# Read csvs
rus_results = pd.read_csv(os.path.join(results_dir, rusmodel_report))
rosrus_results = pd.read_csv(os.path.join(results_dir, rosrusmodel_report))

print(f"RUS results: \n{rus_results}")
print(f"ROS RUS results: \n{rosrus_results}")
```

```python

combined_results = pd.merge(left=rus_results,right=rosrus_results, how="inner", left_on="question__species",
                                right_on="question__species", suffixes=("_x", "_y"))
combined_results= combined_results[combined_results.index < 52] 
print(f"Combined: \n{combined_results}")

combined_results.to_csv(os.path.join(results_dir, fn_combined), sep=',', header=True, index=False)
```

```python
# Compare predictions
rus_predictions = r"res18_adam_epoch1_rus_phase1_15_test_results.csv"
rosrus_predictions = r"res18_adam_epoch1_ros_rus_phase1_14_test_results.csv"

fn_combined_predictions = r"rus_predictions_vs_rosrus_predictions.csv"
```

```python
# Read csvs
rus_predictions_df = pd.read_csv(os.path.join(results_dir, rus_predictions))
rosrus_predictions_df = pd.read_csv(os.path.join(results_dir, rosrus_predictions))

print(f"RUS results: \n{rus_predictions_df}")
print(f"ROS RUS results: \n{rosrus_predictions_df}")
```

```python
combined_predictions_df = pd.merge(left=rus_predictions_df,right=rosrus_predictions_df, how="inner", left_on="image_path",
                                right_on="image_path", suffixes=("_x", "_y"))
print(f"Combined: \n{combined_predictions_df}")

combined_predictions_df.to_csv(os.path.join(results_dir, fn_combined_predictions), sep=',', header=True, index=False)
```
