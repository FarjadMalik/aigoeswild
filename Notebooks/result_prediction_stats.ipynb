```python

predictions_df = pd.read_csv(os.path.join(results_dir, results_file))
# Show initial outputs from the predictions file
print(f"Predictions head:\n{predictions_df.head()}")
print(f"Predictions columns:\n{predictions_df.columns}")
```

```python
# Get true labels and predicted labels from the predictions_df
true_labels = predictions_df.true_label
predicted_labels = predictions_df.predicted_label
print(f"Len True Labels: {len(true_labels)}")
print(f"Len Model Predictions: {len(predicted_labels)}")
```

```python
# Calculate and display statistics
test_accuracy = accuracy_score(true_labels,
                               predicted_labels)
precision = precision_score(true_labels, predicted_labels, average='macro', zero_division=0)
recall = recall_score(true_labels, predicted_labels, average='macro', zero_division=0)
f1 = f1_score(true_labels, predicted_labels, average='macro')


print(f"The accuracy on test set is: {test_accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 score: {f1}")
```

```python
cm = confusion_matrix(true_labels, predicted_labels)
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#array([[1.        , 0.        , 0.        ],
#      [1.        , 0.        , 0.        ],
#      [0.        , 0.33333333, 0.66666667]])

#The diagonal entries are the accuracies of each class
cm.diagonal()
print(f"Confusion Matrix: \n{cm}")
print(f"Confusion Matrix Diagonal: \n{cm.diagonal()}")

```

```python
cr = classification_report(true_labels, predicted_labels, digits=3, zero_division=0, output_dict=True)
prediction_report = pd.DataFrame(cr).transpose()
prediction_report = prediction_report[prediction_report.index.astype('str').str.isnumeric()]
prediction_report['class'] = prediction_report.index
prediction_report['class'] = prediction_report['class'].apply(lambda x: x if x.isnumeric() else 99)
prediction_report['class'] = prediction_report['class'].astype(np.int32)
prediction_report['accuracy'] = cm.diagonal()
print(f"Classification report: \n{prediction_report}")
```

```python
specie_mapping_file = r"C:\Users\mfarj\Documents\ss_data\data_csv\label_to_species_mapping.csv"
species_mapping = pd.read_csv(specie_mapping_file)
print(f"Species mapping: \n{species_mapping}")
```

```python
# Merge with classification report
prediction_report = pd.merge(left=prediction_report,right=species_mapping, how="left", left_on="class",
                             right_on="encoded_species", sort=True)
del prediction_report["encoded_species"]
del prediction_report["class"]
del prediction_report["support"]
print(f"Classification report: \n{prediction_report}")
```

```python
# Read number of images of each specie and join it with predictions
file_train_count_specie = r"C:\Users\mfarj\Documents\ss_data\data_csv\train_count_per_species.csv"
count_species = pd.read_csv(file_train_count_specie)

# Merge with classification report
prediction_report = pd.merge(left=prediction_report,right=count_species, how="inner", left_on="question__species",
                             right_on="question__species")
prediction_report = prediction_report.sort_values('count', ascending=False)
print(f"Classification report: \n{prediction_report}")
```
