```python
import pandas as pd
import os
import numpy as np

from matplotlib import pyplot as plt
# Set the figure size - handy for larger output
plt.rcParams["figure.figsize"] = [6, 12]
```

## Reading classification reports
Here we first input all the reports stored for our models and read them one by one

```python
# Input classification report dir
reports_dir = r"C:\Users\mfarj\Documents\projects\aigoeswild\results"

# Baseline model
file_cr_base = "Baseline_model_test_per_specie_report.csv"
# ROS model phase 1 and 2
file_cr_ros_p1 = "ROS_phase1_10_test_per_specie_report.csv"
file_cr_ros_p2 = "ROS_phase2_7_test_per_specie_report.csv"
# RUS model
file_cr_rus_p1 = "RUS_phase1_14_test_per_specie_report.csv"
file_cr_rus_p2 = "RUS_phase2_14_test_per_specie_report.csv"
```

```python
baseline_cr = pd.read_csv(os.path.join(reports_dir, file_cr_base))
ros_p2_cr = pd.read_csv(os.path.join(reports_dir, file_cr_ros_p2))
ros_p1_cr = pd.read_csv(os.path.join(reports_dir, file_cr_ros_p1))
rus_p1_cr = pd.read_csv(os.path.join(reports_dir, file_cr_rus_p1))
rus_p2_cr = pd.read_csv(os.path.join(reports_dir, file_cr_rus_p2))


# Set species as the index of the reports
baseline_cr = baseline_cr.set_index('Species')
ros_p2_cr = ros_p2_cr.set_index('Species')
ros_p1_cr = ros_p1_cr.set_index('Species')
rus_p1_cr = rus_p1_cr.set_index('Species')
rus_p2_cr = rus_p2_cr.set_index('Species')
```

```python
# Try to put the baseline results upto 3 decimal points
baseline_cr = baseline_cr.round(4)
ros_p2_cr = ros_p2_cr.round(4)
ros_p1_cr = ros_p1_cr.round(4)
rus_p1_cr = rus_p1_cr.round(4)
rus_p2_cr = rus_p2_cr.round(4)

baseline_cr.to_csv(os.path.join(reports_dir, file_cr_base), sep=',', header=True, index=True)
ros_p2_cr.to_csv(os.path.join(reports_dir, file_cr_ros_p2), sep=',', header=True, index=True)
ros_p1_cr.to_csv(os.path.join(reports_dir, file_cr_ros_p1), sep=',', header=True, index=True)
rus_p1_cr.to_csv(os.path.join(reports_dir, file_cr_rus_p1), sep=',', header=True, index=True)
rus_p2_cr.to_csv(os.path.join(reports_dir, file_cr_rus_p2), sep=',', header=True, index=True)
```

```python
file_rosrusp2_cr = r"Balanced(5K)_phase2_14_test_per_specie_report.csv"
rosrusp2_cr = pd.read_csv(os.path.join(reports_dir, file_rosrusp2_cr))
rosrusp2_cr = rosrusp2_cr.set_index('Species')
rosrusp2_cr = rosrusp2_cr.round(4)
rosrusp2_cr.to_csv(os.path.join(reports_dir, file_rosrusp2_cr), sep=',', header=True, index=True)
```

## Plot Baseline results

```python
baseline_cr
```

```python
ax = baseline_cr['F1-score'].plot(kind="barh")
ax.invert_yaxis()

plt.title("Model Baseline -- F1-score per Species")
plt.xlabel("F1-score")
plt.ylabel("Species")
```

## Compare Baseline and ROS (phase2)

```python
# merge ros p2 and baseline and plot bar chart unstacked to see the comparison

comparison_df = pd.merge(left=baseline_cr,right=ros_p2_cr, how="inner", left_on="Species",
                             right_on="Species", suffixes=('_baseline', '_ros'),)
# comparison_df
ax = comparison_df[['F1-score_baseline','F1-score_ros']].plot(kind="barh", stacked=False)
ax.invert_yaxis()

plt.title("Models: Baseline vs ROS")
plt.xlabel("F1-score")
plt.ylabel("Species")
```

# Plot Delta bar chart

```python

comparison_df['delta_F1'] = comparison_df['F1-score_ros'] - comparison_df['F1-score_baseline']
ax = comparison_df['delta_F1'].plot(kind="barh", stacked=True)
ax.invert_yaxis()
ax.set_xlim(-1,1)
print(f"Average Baseline vs ROS delta {comparison_df['delta_F1'].mean()}")
comparison_df['delta_F1_non_zero'] = comparison_df['delta_F1'].replace(0, np.NaN)
print(f"Average Baseline vs ROS phases non zero delta {comparison_df['delta_F1_non_zero'].mean()}")

plt.title("Models: Baseline vs ROS (delta)")
plt.xlabel("F1-score")
plt.ylabel("Species")
```

## Compare Baseline and RUS (phase2)

```python
# merge rus p2 and baseline and plot bar chart unstacked to see the comparison

comparison_base_rus_df = pd.merge(left=baseline_cr,right=rus_p2_cr, how="inner", left_on="Species",
                             right_on="Species", suffixes=('_baseline', '_rus'),)
ax = comparison_base_rus_df[['F1-score_baseline','F1-score_rus']].plot(kind="barh", stacked=False)
ax.invert_yaxis()

plt.title("Models: Baseline vs RUS")
plt.xlabel("F1-score")
plt.ylabel("Species")
```

```python

comparison_base_rus_df['delta_F1'] = comparison_base_rus_df['F1-score_rus'] - comparison_base_rus_df['F1-score_baseline']
ax = comparison_base_rus_df['delta_F1'].plot(kind="barh", stacked=True)
ax.invert_yaxis()
ax.set_xlim(-1,1)

print(f"Average Baseline vs RUS delta {comparison_base_rus_df['delta_F1'].mean()}")
comparison_base_rus_df['delta_F1_non_zero'] = comparison_base_rus_df['delta_F1'].replace(0, np.NaN)
print(f"Average Baseline vs RUS non zero delta {comparison_base_rus_df['delta_F1_non_zero'].mean()}")

plt.title("Models: Baseline vs RUS (delta)")
plt.xlabel("F1-score")
plt.ylabel("Species")
```

## Compare phase 1 and phase 2 of ROS

```python
# merge ros p2 and p1 and plot bar chart unstacked to see the comparison

comparison_ros_phases = pd.merge(left=ros_p1_cr,right=ros_p2_cr, how="inner", left_on="Species",
                             right_on="Species", suffixes=('_ros_p1', '_ros_p2'),)
comparison_ros_phases['delta_F1'] = comparison_ros_phases['F1-score_ros_p2'] - comparison_ros_phases['F1-score_ros_p1']

ax = comparison_ros_phases[['F1-score_ros_p1','F1-score_ros_p2']].plot(kind="barh", stacked=False)
ax.invert_yaxis()

plt.title("Random Oversampling: Phase1 vs Phase2")
plt.xlabel("F1-score")
plt.ylabel("Species")
```

```python

ax = comparison_ros_phases['delta_F1'].plot(kind="barh", stacked=True)
ax.invert_yaxis()
ax.set_xlim(-1,1)
print(f"Average ROS phases delta {comparison_ros_phases['delta_F1'].mean()}")
comparison_ros_phases['delta_F1_non_zero'] = comparison_ros_phases['delta_F1'].replace(0, np.NaN)
print(f"Average ROS phases non zero delta {comparison_ros_phases['delta_F1_non_zero'].mean()}")

plt.title("ROS: Phase1 vs Phase2 (delta)")
plt.xlabel("F1-score")
plt.ylabel("Species")
```
